```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df=pd.read_csv("E:/PYTHON/DAY 4/LAPTOP PRICE - Laptop_Price.csv")

X=df[["RAM_GB"]]
y=df["Price_Rs"]

model = LinearRegression()
model.fit(X, y)

print("Price increase per GB RAM (slope):", round(model.coef_[0], 2))
print("Base price (intercept):", round(model.intercept_, 2))

# Predict price for 16GB RAM
ram=16
predicted_price = model.predict([[ram]])[0]
print(f"Predicted Laptop Price for {ram}GB RAM: ₹{round(predicted_price,2)}")
```

```python
#extra
# Plot actual data and regression line
plt.scatter(X, y, label="Actual data", color="blue")
y_pred = model.predict(X)
plt.plot(X, y_pred, color="red", label="Regression line")
plt.xlabel("RAM (GB)")
plt.ylabel("Laptop Price (₹)")
plt.title("Laptop Price Prediction based on RAM")
plt.legend()
plt.show()
```
